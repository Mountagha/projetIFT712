```python
from sklearn.preprocessing import Imputer
# Ceci est pour gérer les données manquantes dans notre dataset
# imputer calcul dans notre cas ici la mediane pour chaque feature (numerique) afin que ce dernier serve
# de remplacement aux données manquantes
imputer = Imputer(strategy="median")
data_leaf = leaf.drop("id", axis=1)
data_leaf = data_leaf.drop("species", axis=1)
imputer.fit(data_leaf)
imputer.statistics_[:10]
data_leaf.median().values[:10]
```

```python
#Imputer stoque les valeurs calculés dans le tableau statistics_
imputer.statistics_[:10]
```

```python
# Pour vérifier que imputer calcule bien les medianes
import numpy
data_leaf.median().values[:10]
```

```python
# une fois qu'on a fait le calcul on peut appliquer les transformations aux données(ie filling missing datas) 
X_train = imputer.transform(data_leaf)
```

```python
# ceci permet de remettre les données dans le format initial(comme après avoir importé pandas)
leaf_cleaned = pd.DataFrame(X_train, columns=data_leaf.columns)
```

```python
# Les algorithmes de machine learning sont connus pour être plus performants quand ils travaillent avec des données 
# numériques. Convertissons donc nos données textes sous forme de données numériques. Pour cela scikit learn fournit 
# un transformateur pour cette tâche appelé LabelEncoder.
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
leaf_species = leaf["species"]
leaf_species_encoded = encoder.fit_transform(leaf_species)
leaf_species_encoded
```

```python
print(encoder.classes_[:10])
```

```python
# le downside de cette technique notre model pourrait penser que deux catégories ayant des nombres "proches" 
# sont similaires ce qui n'est pas le cas. une solution à ce problème est d'utiliser le one-hot-encoding sur les données 
# déjà labelisés. le fit_transorm de OneHotEncoder attend un 2D alors que les données renvoyés par LabelEncoder 1D
# il faut donc les reshape
from sklearn.preprocessing import OneHotEncoder
encoder = OneHotEncoder()
leaf_species_onehot = encoder.fit_transform(leaf_species_encoded.reshape(-1,1))
leaf_species_onehot
```

```python
#on aurait pu utiliser LabelBinarizer = LabelEncoder + OneHotEncoder
leaf_cleaned.info(verbose=True)
```

```python
leaf_cleaned["species"] = leaf_species_onehot
```
